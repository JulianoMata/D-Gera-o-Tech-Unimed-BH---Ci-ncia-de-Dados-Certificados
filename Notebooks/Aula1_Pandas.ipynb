```python
# Importando a biblioteca 'Pandas'
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JulianoMata/Geracao-Tech-Unimed-BH-Ciencia-de-Dados/main/Dados/gapminder.csv", sep = ";") 
```

```python
# Visualizando as 5 primeiras linhas
df.head()
```

```python
# Total de linhas e colunas
df.shape
```

```python
# Renomeando as colunas
df = df.rename(columns = {"country": "Pais", "continent": "Continente", "year": "Ano", "lifeExp": "Expectativa de Vida", "pop": "Populacao Total", "gdpPercap": "PIB"})
df.head(10).round(2)
```

```python
# Retornando as colunas
df.columns
```

```python
# Tipos de dados de cada coluna
df.dtypes
```

```python
# √öltimas linhas
df.tail(15).round(2)
```

```python
# Retorna informa√ß√µes estat√≠sticas
df.describe().round(2)
```

```python
df["Continente"].unique()
```

```python
Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head(15)
```

```python
Oceania["Continente"].unique()
```

```python
df.groupby("Continente")["Pais"].nunique()
```

```python
df.groupby("Ano")["Expectativa de Vida"].mean().round(2)
```

```python
df["PIB"].mean().round(2)
```

```python
df["PIB"].sum().round(2)
```
